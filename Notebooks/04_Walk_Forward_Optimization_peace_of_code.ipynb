```python
import pandas as pd
import itertools
```

```python
parameters_range = {
    "tp": [0.0035, 0.005, 0.0075, 0.01, 0.0135],
    "sl": [-0.0035, -0.005, -0.0075, -0.01, -0.0135],
}

fixed_parameters = {
    "fast_sma": 60,
    "slow_sma": 120,
    "rsi": 21,
    "cost": 0.0001,
    "leverage": 5
}
```

```python
# Create a list of dictionaries with all the possible combination (ONLY with variable parameters)
keys = list(parameters_range.keys())
keys
```

```python
combinations = list(itertools.product(*[parameters_range[key] for key in keys]))
combinations
```

```python
dict(zip(['tp', 'sl'], ((0.0035, -0.0035))))
```

```python
dictionaries = [dict(zip(keys, combination)) for combination in combinations]
dictionaries
```

```python
# We add the fixed parameters on each dictionary
for dictionary in dictionaries:
    dictionary.update(fixed_parameters)
    
dictionaries
```

```python
df = pd.read_csv("../Data/FixTimeBars/AUDUSD_4H_Admiral_READY.csv", index_col="time", parse_dates=True)
df
```

```python
length_train_set = 5000
pct_train_set = 0.80
train_samples = []
test_samples = []
data = df
anchored = False

# Compute the length of the test set
length_test = int(length_train_set / pct_train_set - length_train_set)

# Initialize size parameters
start = 0
# We don't initialize the end with length_train+length_test because we add it in the loop
end = length_train_set

# We split the data until we can't make more than 2 sub-samples
while (len(data) - end) > 2 * length_test:
    end += length_test

    # If we are at the last sample we take the whole rest to not create a tiny last sample
    if (len(data) - end) < 2 * length_test:
        end = len(data)

    # Fill the samples list depending on if there are anchored or not
    if anchored:
        # We store the train and test set in 2 list to make future computations easier
        train_samples.append(data.iloc[:end - length_test,:])
        test_samples.append(data.iloc[end - length_test: end, :])
    else:
        # We store the train and test set in 2 list to make future computations easier
        train_samples.append(data.iloc[start:end - length_test, :])
        test_samples.append(data.iloc[end - length_test: end, :])

    start += length_test
```

```python
for train, test in zip(train_samples, test_samples):
    print("TRAIN:", train.index[0], train.index[-1], "\tTEST", test.index[0], test.index[-1])
```
