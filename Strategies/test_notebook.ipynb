```python
from datetime import datetime
import MetaTrader5 as mt5
import pandas as pd
```

```python
mt5.initialize()
```

```python
from_date = datetime(2023,1,5)
to_date = datetime(2024,1,5)

ticks = mt5.copy_ticks_from("EURUSD", from_date, 500, mt5.COPY_TICKS_ALL) #copy ticks from gets an integer value for how many ticks we want

df_ticks = pd.DataFrame(ticks)

df_ticks["time"] = pd.to_datetime(df_ticks["time"], unit="s")
```

```python
from_date = datetime(2023,1,5)
to_date = datetime(2024,1,5)

ticks = mt5.copy_ticks_range("EURUSD", to_date, from_date, mt5.COPY_TICKS_ALL) #copy ticks range takes a start and an end

df_ticks = pd.DataFrame(ticks)

df_ticks["time"] = pd.to_datetime(df_ticks["time"], unit="s")
```

```python
from_date = datetime.now()

bars = mt5.copy_rates_from("EURUSD", mt5.TIMEFRAME_H1, from_date, 15) #, mt5.COPY_TICKS_ALL

df_bars = pd.DataFrame(bars)

df_bars["time"] = pd.to_datetime(df_bars["time"], unit="ms")
df_bars
```
